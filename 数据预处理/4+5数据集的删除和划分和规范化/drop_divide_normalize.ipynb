```python
import pandas as pd
diabetes=pd.read_csv('../1确定缺失项/diabetes.csv',engine='python')
diabetes
```

```python
diabetes_drop=diabetes.dropna()
diabetes_dropI=diabetes.drop(['Insulin'],axis=1).dropna()
diabetes_dropS=diabetes.drop(['SkinThickness'],axis=1).dropna()
diabetes_dropIS=diabetes.drop(['Insulin','SkinThickness'],axis=1).dropna()
print(len(diabetes_drop),len(diabetes_dropI),len(diabetes_dropS),len(diabetes_dropIS),)
```

```python
def divide(data):
    train_data=data[0:int(0.6*len(data))]
    test_data=data[int(0.6*len(data)):]
    return train_data,test_data

train_drop,test_drop=divide(diabetes_drop)
train_dropI,test_dropI=divide(diabetes_dropI)
train_dropS,test_dropS=divide(diabetes_dropS)
train_dropIS,test_dropIS=divide(diabetes_dropIS)
print(len(train_drop),len(test_drop))
print(len(train_dropI),len(test_dropI))
print(len(train_dropS),len(test_dropS))
print(len(train_dropIS),len(test_dropIS))
```

```python
def normalize(train_data,test_data):
    mean=train_data.mean()
    std=train_data.std()
    train_data_norm=(train_data-mean)/std
    train_data_norm['Outcome']=train_data['Outcome']
    test_data_norm=(test_data-mean)/std
    test_data_norm['Outcome']=test_data['Outcome']
    print(len(train_data_norm),len(test_data_norm))
    return train_data_norm,test_data_norm

train_norm,test_norm=normalize(train_drop,test_drop)
train_normI,test_normI=normalize(train_dropI,test_dropI)
train_normS,test_normS=normalize(train_dropS,test_dropS)
train_normIS,test_normIS=normalize(train_dropIS,test_dropIS)
```

```python
train_norm.to_csv('train_norm.csv',index=False)
test_norm.to_csv('test_norm.csv',index=False)
train_normI.to_csv('train_normI.csv',index=False)
test_normI.to_csv('test_normI.csv',index=False)
train_normS.to_csv('train_normS.csv',index=False)
test_normS.to_csv('test_normS.csv',index=False)
train_normIS.to_csv('train_normIS.csv',index=False)
test_normIS.to_csv('test_normIS.csv',index=False)
```
